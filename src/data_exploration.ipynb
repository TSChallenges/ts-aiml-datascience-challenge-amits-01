```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt         #  importing some basic libraries to work with the dataset and for visualization



def load_data(filepath):
    """
    Load the dataset into a pandas DataFrame.
    """
    df = pd.read_csv(filepath)
    return df

def check_missing_values(df):
    """
    Check for missing values in the DataFrame.
    """
    missing = df.isnull().values.any()
    print("Missing Values:\n", missing)

def generate_summary_statistics(df):
    """
    Generate summary statistics for key variables.
    """
    summary = df.describe()
    print("Summary Statistics:\n", summary)

def visualize_distributions(df):
    """
    Visualize distributions of age, balance, credit_score, and estimated_salary.
    
    Hint: use sns.histplot() 
    """
    plt.subplot(2, 2, 1)  
    sns.histplot(df['age'], kde=True, color='blue', bins=20)
    plt.title('Distribution of Age')

    plt.subplot(2, 2, 2)  # Grid position (2x2 grid, second subplot)
    sns.histplot(df['balance'], kde=True, color='green', bins=20)
    plt.title('Distribution of Balance')

    plt.subplot(2, 2, 3)  # Grid position (2x2 grid, third subplot)
    sns.histplot(df['credit_score'], kde=True, color='orange', bins=20)
    plt.title('Distribution of Credit Score')

    plt.subplot(2, 2, 4)  # Grid position (2x2 grid, fourth subplot)
    sns.histplot(df['estimated_salary'], kde=True, color='purple', bins=20)
    plt.title('Distribution of Estimated Salary')

    plt.tight_layout()
    plt.show()

def main():
    # Load data
    df = load_data("/workspaces/ts-aiml-datascience-challenge-amits-01/data/bank_churn.csv")

    # Check for missing values
    check_missing_values(df)
    
    # Generate summary statistics
    generate_summary_statistics(df)
    
    # Visualize distributions
    visualize_distributions(df)

if __name__ == "__main__":
    main()
```
